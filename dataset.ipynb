```python
import os

import pandas as pd
from PIL import Image
```

```python
folder = 'data/selfies_dataset/'
files = os.listdir(folder)
metadata = []
images = (Image.open(folder + path) for path in files if path.endswith('.jpg'))
for i, image in enumerate(images):
    image = image.resize((512, 512))
    name =  f'image{i}.png'
    image.save(folder + 'train/' + name)
    metadata.append((name, ''))
```

```python
metadata = pd.DataFrame(metadata, columns=['file_name', 'text'])
metadata
```

```python
metadata.to_csv(folder + 'train/metadata.csv', index=False)
```

```python
from datasets import load_dataset

dataset = load_dataset(".", data_dir=folder)
```
